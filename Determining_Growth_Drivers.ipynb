```python
def get_up_sell_cross_sell(year_churn,year_new,pre_year,next_year):
# remove churned cust from year zero new and remove new from year one
#compare new data based on customer id and check the different from year one and year two
# get distinct customer from year zero and calculate their total in a new df. 
# do the same for year one.  
  previous_year = pre_year
  next_years = next_year

#removing churned users 
  for x in year_churn:
     previous_year.drop(previous_year[previous_year['CUSTOMER_ID'] == x].index ,inplace=True)
#removing new users
  for y in year_new:
    next_years.drop(next_years[next_years['CUSTOMER_ID'] == y].index ,inplace=True)

  current_year = previous_year.iloc[1]['YEAR']
  following_year = next_years.iloc[1]['YEAR']
  print(current_year)
  print(following_year)

  first_year = ''
  second_year =''
  first_year = previous_year.groupby(["CUSTOMER_ID"])['REVENUE'].sum()
  second_year = next_year.groupby(["CUSTOMER_ID"])['REVENUE'].sum()
  #print(first_year)
  # print(first_year.to_list())
  New_frame_one =''
  new_frame_two =''
  final_frame = ''
  result =''
  New_frame_one = (first_year.to_frame())
  new_frame_two =(second_year.to_frame())

  New_frame_one.rename(columns={'REVENUE':current_year}, inplace = True)
  new_frame_two.rename(columns={'REVENUE':following_year}, inplace = True)

  #New_frame_one
  #https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html
  final_frame = [New_frame_one,new_frame_two]
  result = pd.concat(final_frame,axis=1)
  result['SALE_DIFF'] = result[following_year] -result[current_year]
  #https://www.dataquest.io/blog/tutorial-add-column-pandas-dataframe-based-on-if-else-condition/
  result['UpSell_CrossSell'] = np.where(result['SALE_DIFF'] > 0,'Upsell','DownSale')
  return result
```

```python
#Re init variable here because they got overwritten
year_four = Growth_df[Growth_df['YEAR']==2019]
year_five= Growth_df[Growth_df['YEAR']==2020]
year_four_churn = get_churned_cust_list(year_four,year_five)
year_five_new = get_churned_cust_list(year_five,year_four)

year_2019_2020 = get_up_sell_cross_sell(year_four_churn,year_five_new,year_four,year_five)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
year_2019_2020.head()
```
